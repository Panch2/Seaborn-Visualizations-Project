```python
#import seaborn package 

import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
df= sns.load_dataset("tips")
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
sns.lineplot(data=df, x= "total_bill", y="tip")
plt.title("Seaborn Line Plot Exp")
```

```python
sns.relplot(data=df, x= "total_bill", y="tip", kind="scatter")
plt.title("Seaborn Scatter Plot Exp")
```

```python
sns.relplot(data=df, x= "total_bill", y="tip", kind="scatter", col="time")
plt.title("Seaborn Scatter Plot Exp")
```

```python
sns.relplot(data=df, x= "total_bill", y="tip", 
            kind="scatter", col="time",
           hue="smoker")
plt.title("Seaborn Scatter Plot Exp")
```

```python
sns.relplot(data=df, x= "total_bill", y="tip", 
            kind="scatter", col="time",
           hue="smoker", style="smoker")
plt.title("Seaborn Scatter Plot Exp")
```

```python
sns.relplot(data=df, x= "total_bill", y="tip", 
            kind="scatter", col="time",
           hue="smoker", style="smoker",
           size="size")
plt.title("Seaborn Scatter Plot Exp")
```

```python
sns.relplot(data=df, x= "total_bill", y="tip", 
            kind="line", col="time",
           hue="smoker", style="smoker",
           size="size")
plt.title("Seaborn Scatter Plot Exp")
```

```python
#apply a default theme 

sns.set_theme()
sns.relplot(data=df, x= "total_bill", y="tip", 
            kind="scatter", col="time",
           hue="smoker", style="smoker",
           size="size")
```

```python
#Adding a regression line 

sns.set_theme()
sns.lmplot(data=df, x= "total_bill", y="tip", col="time", hue="smoker")
```

```python
#Histrogram 
sns.set_theme()
sns.displot(data=df, x= "total_bill", col="time", kde="True")
```

```python
#Cat Plot 
sns.set_theme()
sns.catplot(data=df, x="day", y="total_bill", kind="violin", hue="smoker",
            split="True")
```

```python
sns.set_theme()
sns.stripplot(data=df, x="total_bill", y= "day", hue="day")
```

```python
sns.set_theme()
sns.stripplot(data=df, x="total_bill", y= "day", hue="size")
```

```python
sns.set_theme()
sns.stripplot(data=df, x="total_bill", y= "day", hue="size", dodge="True")
```

```python
sns.set_theme()
sns.stripplot(data=df, x="total_bill", y= "day", hue="size", 
              dodge=True, jitter=False)
```
